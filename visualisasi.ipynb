```python
import matplotlib.pyplot as plt
```

```python
#grafik kualitas
kualitas_grade_C = [1, 1, 0]
titik_grade_C = [0, 30, 40]

kualitas_grade_B = [0, 1, 1, 0] 
titik_grade_B = [30, 40, 50, 90]

kualitas_grade_A = [0, 1, 1]
titik_grade_A = [50, 90, 100]

plt.title('kualitas')

plt.plot(titik_grade_C, kualitas_grade_C, label = 'Grade C')
plt.plot(titik_grade_B, kualitas_grade_B, label = 'Grade B')
plt.plot(titik_grade_A, kualitas_grade_A, label = 'Grade A')
plt.legend()
plt.show()
```

```python
#grafik harga
harga_murah = [1, 1, 0]
titik_murah = [0, 3, 4]

harga_normal = [0, 1, 1, 0]
titik_normal = [3, 4, 6, 7]

harga_mahal = [0, 1, 1]
titik_mahal = [6, 7, 10]

plt.plot(titik_murah, harga_murah, label = 'Murah')
plt.plot(titik_normal, harga_normal, label = 'Normal')
plt.plot(titik_mahal, harga_mahal, label = 'Mahal')
plt.title('harga')
plt.legend()
plt.show()

```

```python
tidak_layak = [1, 1, 0]
titik_tidak_layak = [0, 40, 50]

cukup_layak = [0, 1, 1, 0]
titik_cukup_layak = [40, 50, 70, 80]

sangat_layak = [0, 1, 1]
titik_sangat_layak = [70, 80, 90]

plt.plot(titik_tidak_layak, tidak_layak, label = 'tidak layak')
plt.plot(titik_cukup_layak, cukup_layak, label = 'cukup layak')
plt.plot(titik_sangat_layak, sangat_layak, label = 'sangat layak')
plt.title('kelayakan')
plt.legend()
plt.show()
```
